```python
# Eliminaci√≥n de variables con baja correlacion segun lo observado en el gr√°fico de calor#
Variable_elimin_train=['Id','MSSubClass','OverallCond','BsmtFinSF2','LowQualFinSF','BsmtHalfBath','3SsnPorch','ScreenPorch','PoolArea','MiscVal','MoSold','YrSold']
train_x=df_train.drop(labels=Variable_elimin_train, axis=1)
train_x.columns
```

```python
# Se reemplaza con la media los valores numericos NaN #
# Se busca las columnas que tengan valores numericos #
col_num_train = train_x.select_dtypes(include=['number'])
col_num_train.columns
```

```python
col_num_test = test_x.select_dtypes(include=['number'])
col_num_test.columns
```

```python
# Se calcula la media y se reemplaza los NaN #
medias_train = col_num_train.mean()
col_num_train = col_num_train.fillna(medias_train)
train_x.update(col_num_train)
train_x.head(10)
```

```python
medias_test = col_num_test.mean()
col_num_test = col_num_test.fillna(medias_test)
test_x.update(col_num_test)
test_x.head(10)
```

```python
# Se verifica que las columnas con valores numericos NaN ya esten reemplazados por la media #
train_x.LotFrontage.isnull().sum()
train_x.MasVnrArea.isnull().sum()
train_x.GarageYrBlt.isnull().sum()
test_x.LotFrontage.isnull().sum()
test_x.MasVnrArea.isnull().sum()
test_x.GarageYrBlt.isnull().sum()
```

El an√°lisis de precios de casas ofrece valiosas perspectivas para compradores, vendedores e inversores inmobiliarios, por ejemplo:

‚úî **La ubicaci√≥n** es fundamental ya que pueden tener precios muy distintos. Varian significativamente segun la ubicacion de barrios, ciudades e incluso vecindarios espec√≠ficos.

‚úî **La tendencia estacional** puede afectar los precios. Las ventas de viviendas tienden a aumentar en primavera y verano, lo que puede llevar a precios mas altos durante esos meses.

‚úî Las casas tienden a aumentar de valor con el tiempo asi que **los inversores pueden aprovechar** esto para obtener ganancias a largo plazo.



üì§ **INSIGHTS CARACTERISTICAS**


‚úî **Las caracteristicas** de una casa, como el tama√±o, el n√∫mero de habitaciones, los acabados y las comodidades, afectan significativamente su precio. Cuanto mas grande o mas lujosa sea una propiedad, generalmente mayor sera su precio de venta.

‚úî **Las renovaciones y mejoras**, como una cocina actualizada o un ba√±o renovado, pueden aumentar el valor de una casa.

‚úî **Las caracteristicas personalizadas**, como piscinas, paisajismo, cuidado y caracteristicas unicas de dise√±o, pueden aumenatr el precio, pero su valor puede varias segun el gusto personal de los compradores.

‚úî **Las tendencias del mercado inmobiliario local** tambien pueden afectar como se valoran las caracteristicas, por ejemplo, en un mercado competitivo, las caractersticas de calidad pueden ser mas valoradas.
