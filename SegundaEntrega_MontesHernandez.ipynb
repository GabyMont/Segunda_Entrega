🟤<b>Contexto Comercial:</b> En el mundo actual, la toma de decisiones informada y respaldada por datos precisos es esencial en el mercado. En este caso, basandonos en el mercado inmobiliario, ya sea que se este buscando comprar una casa como inversión o como un hogar, contar con información sólida es crucial. Es aquí donde nuestro enfoque centrado en los datos y la analítica se centra en una amplia gama de propiedades, incluyendo detalles como la ubicación, el tamaño del lote, la cantidad de habitaciones y baños, la edad de la construcción, las mejoras realizadas y muchas otras características.

🟤<b>Problema Comercial:</b> Cuando una casa se valora incorrectamente debido a una falta de datos precisos sobre sus características, puede llevar a la sobrevaloración, lo que alejará a los posibles compradores, o a la subvaloración, resultando una pérdida de ingresos para el vendedor.

🟤<b>Contexto Analítico:</b> En el pasado, la valoración de las propiedades solía depender en gran medida de la intuición de los expertos y de comparaciones subjetivas. Sin embargo, en la actualidad, la analítica de datos se ha convertido en el pilar fundamental para determinar el precio de las casas de manera precisa y objetiva. Estos factores incluyen el tamaño del lote, la superficie habitable, el número de habitaciones y baños, la edad y estado de la construcción, las mejoras realizadas, la proximidad a servicios y escuelas, y otros atributos específicos. Cada uno de estos elementos es evaluado de manera individual y en combinación para determinar su impacto en el precio final.

```python
# Se reemplaza con la media los valores numericos NaN #
# Se busca las columnas que tengan valores numericos #
col_num_train = train_x.select_dtypes(include=['number'])
col_num_train.columns
```

```python
col_num_test = test_x.select_dtypes(include=['number'])
col_num_test.columns
```

```python
# Se calcula la media y se reemplaza los NaN #
medias_train = col_num_train.mean()
col_num_train = col_num_train.fillna(medias_train)
train_x.update(col_num_train)
train_x.head(10)
```

```python
medias_test = col_num_test.mean()
col_num_test = col_num_test.fillna(medias_test)
test_x.update(col_num_test)
test_x.head(10)
```

✔ **Las renovaciones y mejoras**, como una cocina actualizada o un baño renovado, pueden aumentar el valor de una casa.

✔ **Las caracteristicas personalizadas**, como piscinas, paisajismo, cuidado y caracteristicas unicas de diseño, pueden aumenatr el precio, pero su valor puede varias segun el gusto personal de los compradores.
